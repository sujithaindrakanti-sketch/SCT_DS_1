# Population Distribution Analysis (2024)

## Objective

The objective of this project is to visualize the distribution of population values using a histogram.

The analysis uses the World Bank population dataset and focuses on understanding how population values are distributed across different World Bank entities in 2024.

## Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Load Dataset

The dataset contains population data from 1960 to 2025 for different World Bank entities.

```python
import pandas as pd
df = pd.read_csv( "/content/API_SP.POP.TOTL_DS2_en_csv_v2_267553.csv",skiprows=4)

df.head()
```

## Understanding the Dataset

Before visualization, we inspect the structure of the dataset to understand available columns and data types.

```python
df.shape
```

```python
df.info()
```

## Data Cleaning

The dataset contains an empty column created during export and a year (2025) with no available population values.

These columns are removed before analysis.

```python
# Remove empty columns
df = df.drop(columns=['Unnamed: 70'])

# Remove completely empty years
df = df.drop(columns=['2025'])

df.info()
```

```python
population_2024 = df['2024'].dropna()

population_2024.head()
```

```python
population_2024.describe()
```

## Statistical Summary

The population data shows a large variation between entities.

While most entities have populations below 100 million, a few entities have extremely high population values, causing a right-skewed distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,5))

plt.hist(
    population_2024,
    bins=30
)

plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Population Distribution (2024)")

plt.show()
```

## Import Libraries

The following libraries are used:
- Pandas for data handling
- Matplotlib for visualization
- NumPy for numerical transformations

## Handling Skewed Distribution

The population values vary significantly, ranging from thousands to billions.

Because a few entities have extremely high populations, the raw histogram compresses most values near the lower range.

A logarithmic transformation is applied to improve visualization.

```python
# Apply log transformation

log_population = np.log10(population_2024)

plt.figure(figsize=(10,5))

plt.hist(
    log_population,
    bins=30
)

plt.xlabel("Log10 Population")
plt.ylabel("Frequency")
plt.title("Population Distribution Across World Bank Entities (2024)")

plt.show()
```

## Insights

- The population distribution is not uniform.
- Most World Bank entities have relatively smaller populations.
- A small number of entities have extremely large populations.
- The distribution is right-skewed due to the presence of high population entities.
- Log transformation improves visibility by reducing the impact of extreme values.

## Conclusion

The analysis shows that population sizes vary greatly across World Bank entities in 2024.

The histogram helps identify the overall distribution pattern and highlights the presence of significant differences between small and large population entities.
